```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
# Create dataframe with combined dataset
district_data_df = (school_data_complete)
district_data_df.head(10)
```

```python
# Calculate the number of schools in the DataFrame
school_count = len(district_data_df["school_name"].unique())
print(school_count)

# Calculate the total number of students
student_count = len(district_data_df["Student ID"].unique())
print(student_count)

# Calculate the total budget
budget_total = sum(district_data_df ["budget"].unique())
print(budget_total)

# ----AVG -------
# Calculate the average math score
math_average = district_data_df ["math_score"].mean()
print(math_average)

# Calculate the average reading score
read_average = district_data_df ["reading_score"].mean()
print(read_average)

# ----MATH PASSING-------
math_pass = district_data_df.loc[(district_data_df["math_score"])>=70]
math_total = math_pass.count()["Student ID"]
percent_math_pass = ((math_total/student_count)*100)
print(percent_math_pass)

# ----READ PASSING-------
read_pass = district_data_df.loc[(district_data_df["reading_score"])>=70]
read_total = read_pass.count()["Student ID"]
percent_read_pass = ((read_total/student_count)*100)
print(percent_read_pass)

# ----OVERALL PASSING-------
# Calculate the percentage of students who passed math and reading (% Overall Passing)
# locating overall >=70 scores
math_read_pass = district_data_df.loc[(district_data_df["reading_score"]>= 70) & (district_data_df["math_score"]>= 70)]
#print(math_read_pass)

# locating overall >=70 scores and counting by occurrance on Student ID
math_read_total = math_read_pass.count()["Student ID"]

# using that overall count to calculate percentage
overallper = ((math_read_total/student_count)*100)
print(overallper)
```

```python
# Create a dataframe to hold the above results
summary_dicts = [{"Total Schools" : school_count, 
                  "Total Students" : student_count, 
                  "Total School Budget" : budget_total,
                  "Average Math Score" : math_average, 
                  "Average Reading Score" : read_average,
                  "% Passing Math" : percent_math_pass,
                  "% Passing Reading" : percent_read_pass,
                  "% Overall Passing" : overallper}]
district_summary_df = pd.DataFrame(summary_dicts)
district_summary_df
```
